# Piedra Papel & Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego continúa hasta que un jugador haya acumulado 3 puntos, convirtiéndose en el ganador.

```python
# importo la función random
import random

# defino el número de puntos objetivo
puntos_objetivo = 3

# defino las opciones válidas
opciones = ('piedra', 'papel', 'tijera')

# creo un marcador para sumar los puntos que van sacando los jugadores en cada ronda
puntos_usuario = 0
puntos_ordenador = 0

# creo la variable 'rondas', que se irá sumando
ronda = 1

# defino hasta qué punto se va a repetir el ciclo

while True:
    # quiero que al principio de cada ronda me imprima qué ronda estamos jugando y cuántos puntos lleva cada jugador
    print('**********')
    print('Ronda:', ronda)
    print('El usuario lleva', puntos_usuario, 'puntos')
    print('El ordenador lleva', puntos_ordenador, 'puntos')

    # defino todo lo relativo a la elección del usuario
    eleccion_usuario = input("Elige entre piedra, papel o tijera:")
    # paso la opción del usuario a minúsculas por si se equivoca al escribir
    eleccion_usuario = eleccion_usuario.lower()
    if not eleccion_usuario in opciones:
        print("Opción inválida. Vuelve a intentarlo")
        continue # esto hace que si la elección del usuario es una palabra no válida, vuelva a empezar (que no ejecute todo lo que hay abajo)

    # defino todo lo relativo a la elección del ordenador: que elija aleatoriamente entrer las opciones
    eleccion_ordenador =  random.choice(opciones)

    print('Elección del usuario:', eleccion_usuario)
    print('Elección del ordenador:', eleccion_ordenador)

    # si usuario y ordenador coinciden hay empate
    if eleccion_usuario == eleccion_ordenador:
        print("Empate")

    # opciones si el usuario saca piedra
    elif eleccion_usuario == 'piedra':
        if eleccion_ordenador == 'tijera':
            print('piedra gana a tijera')
            print('¡gana el usuario!')
            puntos_usuario += 1
        # la única opción restante es que el ordenador saque papel
        else:
            print('papel gana a piedra')
            print('¡gana el ordenador!')
            puntos_ordenador += 1

    # opciones si el usuario saca papel
    elif eleccion_usuario == 'papel':
        if eleccion_ordenador == 'piedra':
            print('papel gana a piedra')
            print('¡gana el usuario!')
            puntos_usuario += 1
        # la única opción restante es que el ordenador saque tijera
        else:
            print('tijera gana a papel')
            print('¡gana el ordenador!')
            puntos_ordenador += 1

    #opciones si el usuario saca tijeras
    elif eleccion_usuario == 'tijera':
        if eleccion_ordenador == 'papel':
            print('tijera gana a papel')
            print('¡gana el usuario!')
            puntos_usuario += 1
        #la única opción restante es que el ordenador saque piedra
        else:
            print('piedra gana a tijera')
            print('¡gana el ordenador!')
            puntos_ordenador += 1

    if puntos_usuario == puntos_objetivo:
        print('¡El ganador del juego es el usuario!')
        break

    if puntos_ordenador == puntos_objetivo:
        print('¡El ganador del juego es el ordenador!')
        break

    ronda += 1
```
